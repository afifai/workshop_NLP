```python
import requests

# ambil data dari liputan6.com
r = requests.get('https://www.liputan6.com')
```

```python
from bs4 import BeautifulSoup

# hapus tag HTML
soup = BeautifulSoup(r.text, "html5lib")
print(soup.get_text())
```

```python
artikel = soup.find_all('div', class_='articles--iridescent-list')
artikel[0]
```

```python
# ekstrak judul artikel
artikel[3].find("span", class_="articles--iridescent-list--text-item__title-link-text").get_text()
```

```python
import pandas as pd
list_judul = [] # inisialisasi list kosong
for art in artikel: # looping thd object artikel (kumpulan isi tag articles)
    # ekstrak judul u/ setiap object, dan simpan dalam list
    list_judul.append( art.find("span", class_="articles--iridescent-list--text-item__title-link-text").get_text() )
#print(list_judul)
pd.DataFrame(list_judul, columns=['judul_berita']) # tampilkan
```
