```python
with open('../cerita.txt', encoding='utf-8') as f:
    teks = f.read()
```

```python
import re
# hapus spasi berlebih
teks = re.sub(r' +',' ',teks) # spasi berlebih
teks = re.sub(r'\n+',' ',teks) # \n berlebih
teks = teks.lower() # ubah ke huruf kecil
#print(teks)
```

```python
import os
import nltk
nltk.data.path.append("../nltk_data")
```

```python
from nltk.tokenize import sent_tokenize

# split teks menjadi kumpulan kalimat
kalimat = sent_tokenize(teks) # tokenisasi kalimat
```

```python
from nltk.tokenize import word_tokenize
kata = []
for i in range(len(kalimat)):
    kalimat[i] = re.sub(r"[^a-zA-Z]", " ", kalimat[i])
    kalimat[i] = re.sub(r' +',' ',kalimat[i])
    kata_dlm_kalimat = word_tokenize(kalimat[i])
    kata += kata_dlm_kalimat
```

```python
print("Jumlah kata : ", len(kata))
print("Jumlah kalimat : ", len(kalimat))
```

```python
import pandas as pd
pd.DataFrame(kalimat, columns=['kalimat']).head()
```

```python
# hitung kata unik
jum_kata = {}
for k in kata:
    if k not in jum_kata:
        jum_kata[k] = 1
    else:
        jum_kata[k] += 1
    
```

```python
pd.DataFrame(pd.Series(jum_kata), columns=['jum_kata'])
```
