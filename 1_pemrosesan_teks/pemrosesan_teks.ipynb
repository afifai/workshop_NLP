## Pembersihan Data

```python
import requests

# ambil data dari situs https://www.cnnindonesia.com/nasional
link = 'https://www.cnnindonesia.com/nasional'
r = requests.get(link)
print(r.text)
```

```python
from bs4 import BeautifulSoup

# hapus tag HTML
soup = BeautifulSoup(r.text, "html5lib")
print(soup.get_text())
```

```python
artikel = soup.find_all('article', class_='col_4')
artikel[0]
```

```python
# ekstrak judul artikel
artikel[0].find("h2", class_="title").get_text()
```

```python
import pandas as pd
list_judul = [] # inisialisasi list kosong
for art in artikel: # looping thd object artikel (kumpulan isi tag articles)
    # ekstrak judul u/ setiap object, dan simpan dalam list
    list_judul.append( art.find("h2", class_="title").get_text() )
#print(list_judul)
pd.DataFrame(list_judul, columns=['judul_berita']) # tampilkan
```

## Normalisasi

### Normalisasi case

```python
# Contoh teks
teks = "Mayoritas bos perusahaan di Asia kian sadar akan pentingnya pemanfaatan kecerdasan buatan (artificial intelligence/AI), tetapi menganggap cryptocurrency dan blockchain kurang penting dalam masa depan operasi perseroan."
print(teks)
```

```python
# konversi ke huruf kecil
teks = teks.lower()
print(teks)
```

### Penghapusan tanda baca

```python
import re

# hapus tanda baca menggunakan regex
teks = re.sub(r"[^a-zA-Z0-9]", " ", teks)

# hapus spasi berlebih
teks = re.sub(r' +',' ',teks)
print(teks)
```

### Tokenisasi

```python
# pisahkan antar kata
kumpulan_kata = teks.split()
print(kumpulan_kata)
print("Jumlah kata :", len(kumpulan_kata))
```

### NLTK: Natural Language ToolKit

```python
import os
import nltk
nltk.data.path.append("../nltk_data")
```

```python
# Contoh kata lain
teks = "Menurut Dr. X, Hidup di iklim tropis membuat Air Conditioner (AC) seakan menjadi sebuah kebutuhan pokok tersendiri bagi masyarakatnya. Seperti pula yang terjadi di Indonesia. Tanpa AC, penghuni rumah kerap merasa tak nyaman karena udara yang panas dan lembap."
print(teks)
```

### Tokenisasi kata

```python
from nltk.tokenize import word_tokenize

# split teks menjadi kumpulan kata-kata menggunakan NLTK
kumpulan_kata = word_tokenize(teks)
print(kumpulan_kata)
```

### Tokenisasi kalimat

```python
from nltk.tokenize import sent_tokenize

# split teks menjadi kumpulan kalimat
kalimat = sent_tokenize(teks)
print(kalimat)
```
